```python
from __future__ import division
import keras
from keras.models import model_from_json
import sys, os
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
import warnings
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.losses import categorical_crossentropy
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from keras import regularizers
from tensorflow.keras.regularizers import l1
from keras.optimizers import Adam
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

json_file = open("../input/feardataset/CNNFear .json", 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("../input/feardataset/CNNFear .h5")
np.set_printoptions(threshold=np.inf)
predicted_value = []
true_value = []
x = np.load('../input/feardataset/ftestFear .npy')
y = np.load('../input/feardataset/ltestFear.npy')
y1= loaded_model.predict(x)
y2 = y1.tolist()
y3 = y.tolist()
count = 0
detect = 0
packet = 0
for i in range(len(y)):
    y4 = max(y2[i])
    y5 = max(y3[i])
    predicted_value.append(y2[i].index(y4))
    true_value.append(y3[i].index(y5))

    if(y2[i].index(y4) == 1):
        if(y2[i].index(y4) == y3[i].index(y5)):
            count += 1 
    if(y2[i].index(y4) == 0):
        if(y2[i].index(y4) == y3[i].index(y5)):
            packet += 1

    if(y2[i].index(y4) == 0):
        if(y2[i].index(y4) != y3[i].index(y5)):
            detect += 1

Fear_Faces = count
Fear_Faces_Wrong = detect
Not_Fear_Faces = packet
print("Fear Faces Detected :"+str(Fear_Faces))
print("Fear Faces Detected incorrectly:"+str(Fear_Faces_Wrong))
print("Not Fear Faces Detected:"+str(Not_Fear_Faces))

Con_Matrix = confusion_matrix(true_value, predicted_value)
labels = ['Not Fear', 'Fear']
title = 'Confusion matrix'
print(Con_Matrix)

plt.imshow(Con_Matrix, interpolation = 'nearest', cmap = plt.cm.Blues)
plt.title(title)
plt.colorbar()

tick_marks = np.arange(len(labels))
plt.xticks(tick_marks, labels, rotation=45)
plt.yticks(tick_marks, labels)
#as part of the format of the text of the confusion matrix, the output is a decimal Integer. Outputs the number in base 10.
formD = 'd'

threshold = Con_Matrix.max() / 2.

for i, j in itertools.product(range(Con_Matrix.shape[0]), range(Con_Matrix.shape[1])):
    plt.text(j, i, format(Con_Matrix[i, j], formD),
            horizontalalignment = "center",
            color = "white" if Con_Matrix[i, j] > threshold else "black")

plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
```
