```python
#necesarry libaries imported
from __future__ import division
import keras
from keras.models import model_from_json
import sys, os
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
import warnings
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

#loads JSON and creates model
json_file = open('CNN.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
#applies the weights to the created model
loaded_model.load_weights("CNN.h5")

#evaluate loaded model onto test data
predicted_value = []
true_value = []
count = 0 
#load in saved test data
x = np.load('ftest.npy')
#load saved labels
y = np.load('ltest.npy')

#range(len()) is used to access each item in the list 'y' by index within the for-loop
for i in range(len(y)):
    
    #loaded_model.predict() predicts the label of the dataset 'x' and returns the learnt label for each object within the array and .tolist() converts the string to a list
    y1 = (loaded_model.predict(x).tolist()[i])
    #.tolist() converts array 'y' into a list
    y2 = (y.tolist()[i])
    #.index() searches an element in the list and then adds the index value returned to the end of the 'predicted_value' list using .append()
    predicted_value.append(loaded_model.predict(x).tolist()[i].index(y1))
    #.index() searches an element in the list and then adds the index value returned to the end of the 'true_value' list using .append()
    true_value.append(y.tolist()[i].index(y2))
    
    #if statement, shows that when predicted_value label is the same as true_value label to add '1' to 'count'
    if((loaded_model.predict(x).tolist())[i].index(y1) == (y.tolist())[i].index(y2)):
        count+=1

acc = (count/len(y))*100
print("Test Set Accuracy :"+str(acc)+"%")
```

```python
#sets true_value on the y axis, and predicted_value on the x axis
Con_Matrix = confusion_matrix(true_value, predicted_value)
#labels for each of the index numbers: 0,1,2,3,4,5,6 respectively
labels = ['Angry', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']
#give the confusion matrix a title
title = 'Confusion matrix'
#plot confusion matrix, choosing a blue colour map so its easier on the eyes.
plt.imshow(Con_Matrix, interpolation = 'nearest', cmap = plt.cm.Blues)
plt.title(title)
plt.colorbar()

mark = np.arange(len(labels))
#sets the current tick locations and labels of the x-axis. Rotate the tick labels by 45 degrees
plt.xticks(mark, labels, rotation = 45)
#sets the current tick locations and labels of the y-axis.
plt.yticks(mark, labels)

#as part of the format of the text of the confusion matrix, the output is a decimal Integer. Outputs the number in base 10.

threshold = Con_Matrix.max() / 2.

for i, j in itertools.product(range(Con_Matrix.shape[0]), range(Con_Matrix.shape[1])):
    plt.text(j, i, format(Con_Matrix[i, j], 'd'),
            horizontalalignment = "center",
            color = "white" if Con_Matrix[i, j] > threshold else "black")
    
#plots Confusion Matrix with True against Predicted
plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.tight_layout()
plt.show()
```
