```python
#importing necessary libraries
import keras
import sys, os
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
import warnings
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.losses import categorical_crossentropy
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from keras import regularizers
from tensorflow.keras.regularizers import l1
import matplotlib.pyplot as plt
#reading the .csv file containing the dataset information using a pandas function 'pd.read_csv()'' with a dataframe called 'data'
data = pd.read_csv("../input/ferdataset/icml_face_data.csv")
#extracting the pixel information from the pixel column in the dataset, and converting to a list
data_elements = data[' pixels'].tolist()

#retrieving pixel features for training
A = []
for data_sequence in data_elements:
    #'.split()' function splits the pixel value strings extracted from the dataset into a list
    face_features = [int(pixel) for pixel in data_sequence.split(' ')]
    #converts input list 'face_features' into an array 
    #'.reshape' gives a new shape to the array without changing the data
    face_features = np.asarray(face_features).reshape(48, 48)
    #'.append' is a function that allows for an item to be added to the end of a list
    #the item added to the end of the list are the 'face_features' list 
    #'face_features' is converted into a float32 type using the function'.astype'
    A.append(face_features.astype('float32'))
#Converted the list 'A' into an array
A = np.asarray(A)
#expanding the shape of the array by adding another dimension using 'np.expand_dims()'
A = np.expand_dims(A, -1)

#retrieving emotion labels from the 'emotion' column in the dataset using pandas function pd.get_dummies()
#'pd.get_dummies()' allows for the dataframe 'data' with the column 'emotion' to be replaced with one-hot encodings and then to interpret the ouput as a matrix using '.as_matrix()'
B = pd.get_dummies(data['emotion']).as_matrix()

#storing features and labels, respectively,saving into a binary NumPy file in .npy format
np.save('features', A)
np.save('labels', B)
```
